1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
# importando bibliotecas

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
# carregando arquivo e construindo DataFrame

arquivo = pd.read_csv('PREVISAO_DE_RENDA.csv')
df_previsao = pd.DataFrame(arquivo)


# informações sobre os dados

df_previsao.dtypes
````

````python
# separando dados númericos para consulta e colunas dispensáveis

df_consulta = df_previsao.select_dtypes(exclude=['object'])
df_consulta = df_consulta.drop('Unnamed: 0', axis=1, errors='ignore')
df_consulta = df_consulta.drop('index', axis=1, errors='ignore')

df_consulta.columns
````

````python
# correlação entre as variáveis

df_consulta.corr()
````

````python
# Comparação das variável "posse_de_imovel" com "renda"

ax = sns.pointplot(x="posse_de_imovel", y="renda", 
                   data=df_consulta, dodge=True, ci=90)

plt.title('Posse de Imóvel x Renda')
plt.show()
````

````python
# Comparação das variável "posse_de_veiculo" com "renda"

ax = sns.pointplot(x="posse_de_veiculo", y="renda", 
                   data=df_consulta, dodge=True, ci=90, color='orange')

plt.title('Posse de Veículo x Renda')
plt.show()
````

````python
# Plotando as duas variáveis x renda

ax = sns.pointplot(x="posse_de_imovel", y="renda", data=df_consulta, dodge=True, ci=90)
ax = sns.pointplot(x="posse_de_veiculo", y="renda", data=df_consulta, dodge=True, ci=90, color='orange')


plt.title('Posse de Imóvel e Posse de Veículo comparadas à Renda')
plt.xlabel('')
plt.ylabel('')
plt.show()
````

1) Alguma das variáveis parece discriminar *renda* de forma estatisticamente significante?
- **Resposta:** Sim, a variável **'posse_de_veiculo'** se relaciona positivamente para o aumento da variável **'renda'**. A variável **'posse_de_imovel'** contra intuitivamente impacta menos o aumento da variável **'renda'**.
